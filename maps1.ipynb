```python
import numpy as np
import matplotlib.pyplot as plt
```

 The logistic map is a widely known map used to model predator-prey interactions over discrete time steps.
 $R$ measures the prey growth rate vs the predator hunting rate, $x_n$ is the population of prey at time $n$.
 $ x_{n+1} = R \cdot x_n \cdot (1 - x_n)$

```python
def logistic_map(x0, r, n):
    x = x0
    while n>0:
        x = logistic_iterator(x,r)
        n -= 1
    return x    

def logistic_iterator(x, r):
    return r * x * (1-x)
```

```python
def logistic_map(x0, r, n):
    x = x0
    xs = [x0]
    while n>0:
        x = logistic_iterator(x,r)
        xs.append(x)
        n -= 1
    return np.array(xs)

def logistic_iterator(x, r):
    return r * x * (1-x)
```

## Question 1

```python
x0 = 0.2
y0 = 0.200001
n = 200
```

```python
r = 2

fig, ax = plt.subplots(1,1,figsize=(8,4))

#ax.set_ylim([0,1])
ax.set_xlabel("n", fontsize=15)
ax.set_ylabel("$|x_n - y_n|$", fontsize=15)

xs = logistic_map(x0,r,n)
ys = logistic_map(y0,r,n)
ax.scatter(list(range(n+1)),np.abs(xs-ys), s=3)
plt.show()
```
